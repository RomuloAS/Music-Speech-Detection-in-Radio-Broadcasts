### Music prediction

Load arff files to pandas dataframe and predict music or no music.

```python
### Create new folder for predictions
path = "../Prediction/Music/"
pathlib.Path(path).mkdir(parents=True, exist_ok=True)

for f in music_files: 
    with open(f) as arff:
        
        ### Load data
        print("Loading data from arff file:", f.split("/")[-1])
        data, meta = loadarff(arff)
        df = pd.DataFrame(data)
        
        ### Convert to stack format and predict
        S = stack.transform(df[cols])
        pred = ensemble.predict_proba(S)
        
        ### Eliminate prediction lower than the threshold
        music_prediction = []
        for p in pred[:,0]:
            if p >= 0.8:
                music_prediction.append("1")
            else:
                music_prediction.append("0")
        
        ### Write file with predictions
        name = f.split("/")[-1]
        name = ".".join(name.split(".")[:-1]) + ".txt"
        print("saving prediction file:", name, end="\n\n")
        with open(path + name, "w") as pred_file:
            pred_file.write("\n".join(music_prediction))
```
